# Google_Scholar_Crawler


### importing necessary libraries

```python
import pandas as pd
import numpy as np
from pandas import read_excel
import math
import requests
from bs4 import BeautifulSoup
import re
import time


```

### Function Definations

```python
def enable_debug_mode(debug_bool):
    if debug_bool == True:
        web_site = 'http://127.0.0.1:5000'
        base_url="http://127.0.0.1:5000/"
    else:
        web_site = 'https://scholar.google.com'
        base_url="https://scholar.google.com/citations?hl=en&view_op=search_authors&mauthors="
    return web_site, base_url

web_site, base_url=enable_debug_mode(False)

file_name = 'newFile.xlsx' 
df = read_excel(file_name)

def download_mainpage(name, surname):
    r=requests.get(base_url + name + "+" + surname)
    print(r.status_code)
    return r.text

def download_subpage(link):
    r1=requests.get(web_site + link)
    return r1.text

def name_surname():
    all=[]
    for i in range(len(df)):
        name = df.iloc[i][1]
        surname = df.iloc[i][0]
        if isinstance(name, str):
            all.append([name, surname])
        else:
            break
    return all

def find_and_extract_data(soup):
    table=[]
    for each_td in soup.find_all("td",{'class':'gsc_rsb_std'}):
        temp=each_td.get_text()
        table.append(temp)
    return table
```

### Main Function

```python
def main():
    
    all=name_surname()
    size_db = len(name_surname())
    test_df=pd.DataFrame()
    for i in range(size_db):
        name=name_surname()[i][1]
        surname=name_surname()[i][0]
        time.sleep(1)
        soup = BeautifulSoup(download_mainpage(name, surname), 'html.parser')
        result=soup.find("h3",{'class':'gs_ai_name'})
       
        if result is None:
            temp=["-","-","-","-","-","-",name,surname]
            print(temp)
            test_df = test_df.append(pd.DataFrame(np.array(temp).reshape(-1,len(temp)), 
                                   columns =['Citation All','Citation Since 2015','h-index All','h-index Since 2015','i10-index All','i10-index Since 2015','First_Name','Last_Name'],
                                  ) )
         
            continue
        else:
            link= result.find('a', href = re.compile(r'[/]([a-z]|[A-Z])\w+')).attrs['href']
            #print("link",link)
            soup = BeautifulSoup(download_subpage(link), 'html.parser')
            #print("soup",soup)
            Data=[]
            
            Data=find_and_extract_data(soup)
            Data.append(name)
            Data.append(surname)
         
            
            test_df = test_df.append(pd.DataFrame(np.array(Data).reshape(-1,len(Data)), 
                                   columns =['Citation All','Citation Since 2015','h-index All','h-index Since 2015','i10-index All','i10-index Since 2015','First_Name','Last_Name'],
                                  ) )
        
            test_df.to_csv('Result_full.csv', index=False)
    


    
main()

#we have total 6 important columns citation all, citation since 2015, h-index All, h-index Since 2015, i10-index All, i10-index Since 2015
```
