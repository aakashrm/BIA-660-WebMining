```python
import bs4
import requests
import re
import pandas as pd
import numpy as np

result = requests.get("https://en.m.wikipedia.org/wiki/List_of_Internet_top-level_domains")
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(result.content, 'html.parser')
# print(soup.prettify()[:3000])
```

```python
soup
```

```python
table = soup.find('table', {'class': ['wikitable']})
urls = ['https://example' + row.find('td').string for row in table.findAll('tr', {'valign': 'top'})]
print(urls)
```

```python
status_codes = []
for link in urls:
#     urllist = link.get_text()
#     url = "http://example" + urllist
#     vdurl.append(url)
    try:
        res = requests.head(link)
        status_codes.append(res.status_code)
    except:
        status_codes.append('Invalid')
```
